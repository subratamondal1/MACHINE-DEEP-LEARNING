## Histogram with pure python

```python
a = ["a", "a", "a", "a", "a", "b", "b", "b", "b", "c", "c", "c"]

def count_elements(seq) -> dict:
    hist = {}
    
    for i in seq:
        hist[i] = hist.get(i, 0) + 1
    return hist

counted = count_elements(a)
print(counted)
```

### Counter: the same code above can be achieved with this class

```python
from collections import Counter

recounted = Counter(a)
print(recounted)
```

#### Verify

```python
counted.items() == recounted.items()
```

### Histogram

```python
import random
random.seed(1)

def histogram(seq) -> None:
    counted = count_elements(seq)
    
    for k in sorted(counted):
        print(f"{k} {'+' * counted[k]}")
    
histogram(a)
```

## Histogram with Numpy

```python
import numpy as np
np.random.seed(444)
np.set_printoptions(precision=3)

d = np.random.laplace(loc=15,scale=3,size=500)

# create bins 
hist_frequency, bin_edges = np.histogram(d)

print(hist_frequency, "\n")
print(bin_edges)
```

* the values inside **hist_frequency** represents the frequency of each bins
* the values inside **bin_edges** represents the start and end of a bin i.e
    * the first bin starts from 2.11 to 5.874
    * the second bin starts from 5.874 to 9.638 & so on...

## Histogram with Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(444)
np.set_printoptions(precision=3)

d = np.random.laplace(loc=15,scale=3,size=500)

hist_freq, bins_edges, patches = plt.hist(x = d, bins= "auto", color="#0504aa", alpha = 0.75, rwidth=0.85)

plt.grid(axis="y", alpha=0.75)
plt.title("Histogram")
plt.xlabel("values")
plt.ylabel("frequency")
max_frequency = hist_freq.max()
plt.ylim(top=np.ceil(max_frequency/10)*10 if max_frequency % 10 else max_frequency+10)

plt.text(20,52, "subrata\nmondal")
plt.show()
```

## Histogram with Pandas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

commute = pd.Series(np.random.gamma(1, scale=10, size=1000) ** 1.5)

commute.plot.hist(grid=True, bins=20, rwidth=0.85, color = "#607c8e")

plt.grid(axis="y", alpha = 0.7)
plt.title("Histogram")
plt.xlabel("commute time")
plt.ylabel("counts")

plt.text(215,52, "subrata\nmondal")
plt.show()
```

## Kernel Density Estimation: KDE

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

means = 10, 20
std_devs = 4, 2

dist = pd.DataFrame(
    np.random.normal(loc=means,scale=std_devs,size=(1000,2)), 
    columns=["a","b"]
)

print(dist.agg(["min","max","mean","std"]).round(3))

fig, ax = plt.subplots()
dist.plot.kde(ax=ax, legend = False, title = "Histogram A vs B")
dist.plot.hist(ax=ax, density = True)

ax.set_ylabel("probability")
ax.grid(axis="y")
ax.set_facecolor("#d8dcd6")
plt.show()
```

## Pandas tools to increase productivity

### pd.cut()

```python
ages = pd.Series([1,1,3,5,8,10,12,15,18,18,19,20,25,30,40,51,52])

bins = [0,10,13,18,21,np.inf]

labels = ["child", "preteen", "teen", "military age", "adult"]

groups = pd.cut(ages, bins=bins,labels=labels)

groups.value_counts()
```
